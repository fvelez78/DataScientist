## COMBINANDO DATAFRAMES

```python
# Concatenacion de dataframes
dt09=pd.DataFrame({'Pais':['Colombia','Per√∫','Ecuador','Bolivia','Venezuela','Paraguay','Nicaragua'],
                   'Capital':['Bogot√°','Lima','Quito','La Paz','Caracas','Asuncion','Managua']})
dt10=pd.DataFrame({'Pais':['Colombia','Per√∫','Ecuador','Bolivia','Venezuela','Paraguay','El salvador'],
                   'Pob (Mill)':[43,8,6,5,29,11,4]})
dt11=pd.DataFrame({'Pais':['Brasil','Argentina','Chile','Uruguay','Panama'],
                   'Capital':['Brasilia','Buenos aires','Santiago de chile','Montevideo','Panama']})
```

```python
# axis indica la dimension en la cual se realiza la concatenacion 0= filas 1= columnas
dt12=pd.concat([dt09,dt11],axis=0)
dt12
```

## Combinacion de dataframes con MERGE

```python
# Combinacion de dos dataframe
dt13=dt09.merge(dt10)  # POr defecto realiza un inner join
dt13
```

```python
# La instrccion on indica la variable sobre la cual se realiza el merge, pero esta deber√° se comun a los dos dataframes.
dt14=dt09.merge(dt10,on='Pais') 
dt14
```

```python
# En caso que los campos llave no tengan el mismo nombre, se debe especificar por separado.
dt15=pd.DataFrame({'Paises':['Colombia','Per√∫','Ecuador','Bolivia','Venezuela','Paraguay','El salvador'],
                   'Pob (Mill)':[43,8,6,5,29,11,4]})
dt16=dt09.merge(dt15,left_on='Pais',right_on='Paises')
dt16
```

```python
# En caso que algunos valores de los campos llave sean NA, entonces solo se traen los registros con llave valida
dt17=pd.DataFrame({'Paises':['Colombia',np.nan,'Ecuador','Bolivia','Venezuela','Paraguay','El salvador'],
                   'Pob (Mill)':[43,8,6,5,29,11,4]})
dt18=dt09.merge(dt17,left_on='Pais',right_on='Paises')
# La salida ya no contiene a Per√∫
dt18
```

```python
# Para otros tipos de join se debe usar el comando how asi:
# En caso que los campos llave no tengan el mismo nombre, se debe especificar por separado.
dt15=pd.DataFrame({'Paises':['Colombia','Per√∫','Ecuador','Bolivia','Venezuela','Paraguay','El salvador'],
                   'Pob (Mill)':[43,8,6,5,29,11,4]})
dt19=dt09.merge(dt15,left_on='Pais',right_on='Paises',how='left')
dt19
```

```python
dt20=dt09.merge(dt15,left_on='Pais',right_on='Paises',how='right')
dt20
```

En el caso de hacer combinacion de Dataframes con JOIN solo se debe tener encueta que por defecto esta instrucci√≥n realiza la combinacion usando los index de cada dataframe.
